```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
sns.barplot(x='Survived',y='Pclass',data=df)
plt.show()
```

```python
sns.barplot(x='Survived',y='Pclass',hue='Sex',data=df)
plt.show()
```

```python
sns.lineplot(x='Age',y='Survived',data=df)
plt.title('age vs survival')
plt.show()
```

```python
data = sns.load_dataset("iris")
 
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

plt.subplot(121)
graph()

plt.subplot(122)
graph()
 
plt.show()
```

```python
#relational plots
sns.relplot(x='Age',y='Survived',data=df)
plt.show()
```

```python
#scatter plots
sns.scatterplot(x='Pclass',y='Survived',data=df)
plt.show()
```

```python
#relational plots
sns.boxplot(x='Age',y='Survived',data=df)
plt.show()
```
