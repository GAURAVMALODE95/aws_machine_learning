```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day48-simple-linear-regression/placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_tr
```

```python
x_te
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr
```

```python
lr.fit(x_tr,y_tr)
```

```python
x_te
```

```python
y_te
```

```python
lr.predict(x_te.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_tr,lr.predict(x_tr),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
b = lr.intercept_
```
