```python
#matplotlib is used for data visuliastion
import numpy as np
from matplotlib import pyplot as plt 
```

# 1.line plot

```python
z=np.arange(1,20)
z
```

```python
y = 2*z  
y
```

```python
plt.plot(z,y)
plt.show()
```

```python
plt.plot(z,y)
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(z,y,color='r',linestyle=':',linewidth=4)
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=np.arange(1,20)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':',linewidth=4)
plt.plot(x,y2,color='g',linewidth=4)
plt.title('two line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.grid(True)#makes grid in plot
plt.show()
```

# BAR PLOT

```python
 #used for catogerical data
student={"roshan":56,"chandra":34,"shekhar":66,"chaudhari":99}
```

```python
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```

```python
plt.title('moon plot')
plt.xlabel('peoples')
plt.ylabel('age')
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```

```python
plt.title('moon plot')
plt.xlabel('peoples')
plt.ylabel('age')
names=list(student.keys())
values=list(student.values())
plt.barh(names,values,color='y')
plt.show()
#same bar plot but horizantal
```
