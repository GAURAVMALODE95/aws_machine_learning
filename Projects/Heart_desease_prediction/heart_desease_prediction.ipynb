CP (Chest Pain): This column describes the type of chest pain people experienced. It can be categorized into different levels like mild, moderate, or severe chest pain.

Trestbps (Resting Blood Pressure): This column shows the resting blood pressure of individuals. It measures the force of blood against the walls of arteries when they are at rest.

Chol (Cholesterol): This column represents the cholesterol level in the blood. Cholesterol is a type of fat in the blood and can be high or low.

Fbs (Fasting Blood Sugar): This column indicates the fasting blood sugar level. It tells us whether an individual has high blood sugar or not.

Restecg (Resting Electrocardiographic Results): This column describes the results of an electrocardiogram (ECG) when a person is at rest. It helps in assessing heart health.

Thalach (Maximum Heart Rate Achieved): This column shows the highest heart rate a person achieved during a cardiac stress test or exercise.

Exang (Exercise-Induced Angina): This column tells us whether a person experienced angina (chest pain) during exercise.

Oldpeak (ST Depression): This column measures the ST segment depression on an ECG during exercise relative to rest. It's related to heart activity.

Slope: This column represents the slope of the peak exercise ST segment. It's another ECG-related measure.

Ca (Number of Major Vessels Colored by Fluoroscopy): This column indicates the number of major blood vessels that were seen with dye during a fluoroscopy procedure.

Thal (Thalassemia): This column describes a blood disorder called thalassemia. It can have different categories like normal, fixed defect, or reversible defect.

```python
from sklearn.decomposition import PCA


# Apply PCA to reduce the dimensionality
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(df)

# Create a scatter plot of the reduced data
plt.scatter(X_reduced[:, 0], X_reduced[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of the Dataset')

plt.show()
```

```python
x = df.drop(columns=['target'])
y=df['target']
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'X' is your feature data and 'y' is your target variable
x_tr, x_te, y_tr, y_te = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
logistic= LogisticRegression(random_state=42)
rf = RandomForestClassifier(n_estimators=20,max_depth=10)
```

```python
voting_clf = VotingClassifier(estimators=[
    ('logistic', logistic),
    ('random_forest', rf)
], voting='soft',weights=[1, 2])
```

```python
voting_clf.fit(x_tr,y_tr)
```

```python
y_pred = voting_clf.predict(x_te)
```

```python
accuracy = accuracy_score(y_te, y_pred)
```

```python
print(accuracy*100)
```
