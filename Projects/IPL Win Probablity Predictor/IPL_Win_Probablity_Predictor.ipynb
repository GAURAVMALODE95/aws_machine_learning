```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]#only teams will be displayed which are present in teams list
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df['dl_applied'].value_counts()#only 15 matches on which dls is applied
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
match_df
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df
```

```python
delivery_df.shape
```

```python
#after each ball how much runs left and how many balls left
delivery_df.groupby('match_id').cumsum()['total_runs_y']
#cumsum=cummelitive summation
```

```python
delivery_df['current_runs']=delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_runs']
```

```python
delivery_df
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
#current run rate=runs/overs
delivery_df['crr']=(delivery_df['current_runs']*6)/(120-delivery_df['balls_left'])
```

```python
#rrr
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
from sklearn.model_selection import train_test_split
x=final_df.iloc[:,0:-1]
y=final_df['result']
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_tr
```

```python
y_te
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])

]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step 1',trf),
    ('step 2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_tr,y_tr)
```

```python
y_pred=pipe.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_te,y_pred)
```
